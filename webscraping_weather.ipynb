### simple web scrapper to get 1 - 3 day forecast from weather-forecast.com

```python
import re
import urllib2
import requests

# ex url
# https://www.weather-forecast.com/locations/Seattle/forecasts/latest
url_prefix = 'https://www.weather-forecast.com/locations/'
url_suffix = '/forecasts/latest'
```

```python
city = 'Seattle'
url = url_prefix + city + url_suffix

r = requests.get(url)
forecast_s = r.text
forecast_s
```

```python
search_start = 'Seattle 1 &ndash; 3 Day Weather Forecast Summary:</b><span class="read-more-small"><span class="read-more-content"> <span class="phrase">'
search_end = '</span></span></span></p><div class="forecast-cont"><div class="units-cont">'

start_pos = re.search(search_start, forecast_s).end()
end_pos = re.search(search_end, forecast_s).start()

forecast_s[start_pos:end_pos]
```
