```python
%matplotlib inline

import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
#basic
x = [1,2,3,4]
y = [3,8,10,25]

plt.title('Rain in Dec')
plt.plot(x, y)

plt.show()
```

```python
#messing with properties
x1 = [3,7,8,12]
y1 = [5,13,2,8]
x2 = [1,2,3,4]
y2 = [3,8,10,25]
x3 = [0,4,6,12]
y3 = [12,5,8,2]

fig = plt.figure()
fig.patch.set_facecolor('green')

fig1 = fig.add_subplot(2,1,1, facecolor='black')
fig1.plot(x1, y1, 'red', linewidth=1.5)
fig1.plot(x2, y2, 'w', linewidth=3)

fig1.tick_params(color='white')
fig1.spines['bottom'].set_color('r')
fig1.spines['left'].set_color('r')
fig1.set_title('TRASH GRAPH 1', color='w')
fig1.set_xlabel('foo', color='w')
fig1.set_ylabel('bar', color='w')


fig2 = fig.add_subplot(2,1,2, facecolor='black')
fig2.plot(x3, y3, 'w', linewidth=1.5)

fig2.tick_params(color='white')
fig2.spines['bottom'].set_color('r')
fig2.spines['left'].set_color('r')
fig2.set_title('TRASH GRAPH 2', color='w')
fig2.set_xlabel('foo', color='w')
fig2.set_ylabel('bar', color='w')

plt.show()
```

```python
# 3d scatter
from mpl_toolkits.mplot3d import axes3d

x = [3,7,8,12]
y = [5,13,2,8]
z = [1,2,3,4]

fig = plt.figure()
fig1 = fig.add_subplot(1,1,1, projection='3d')

fig1.scatter(x,y,z, marker='o')

plt.show()
```

```python
# 3d wireframe
from mpl_toolkits.mplot3d import axes3d

x,y,z = axes3d.get_test_data(.05)

fig = plt.figure()
fig1 = fig.add_subplot(1,1,1, projection='3d')

fig1.plot_wireframe(x,y,z, rstride=10, cstride=10)
```
